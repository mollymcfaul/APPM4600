```python
import matplotlib.pyplot as plt
import numpy as np
import math
from numpy.linalg import inv 
import numpy.linalg as la
```

# PreLab

```python
def find_xeval(xeval, xint):
    
    for x in xint:
        if x != 0:
            ind = np.where(np.logical_and(xeval<=x, xeval>=(x-1)))
    n = len(ind)
    return [ind, n]

def line_maker(x0,fx0,x):
    
    x1 = x0[0]
    x2 = x0[1]
    
    y1 = fx0[0]
    y2 = fx0[1]
    
    yval = ((y2 - y1)/(x2-x1))*(x - x2) + y2
    return yval
```

# 3.1 and 3.2

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from numpy.linalg import inv 


def driver():
    
    f = lambda x: 1/(1+(10*x)**2)
    a = -1
    b = 1
    
    ''' create points you want to evaluate at'''
    Neval = 100
    xeval =  np.linspace(a,b,Neval)
    
    ''' number of intervals'''
    Nint = 10
    Nint1 = 7
    Nint2 = 18
    
    '''evaluate the linear spline'''
    yeval = eval_lin_spline(xeval,Neval,a,b,f,Nint)
    yeval1 = eval_lin_spline(xeval,Neval,a,b,f,Nint1)
    yeval2 = eval_lin_spline(xeval,Neval,a,b,f,Nint2)
    ''' evaluate f at the evaluation points'''
    fex = np.zeros(Neval)
    for j in range(Neval):
      fex[j] = f(xeval[j]) 
      
    
    plt.figure()
    plt.plot(xeval,fex,'ro-',label='true')
    plt.plot(xeval,yeval1,'cs-',label='approx, N = 7')
    plt.plot(xeval,yeval,'gs-',label='approx, N = 10')
    plt.plot(xeval,yeval2,'bs-',label='approx, N =18')
    plt.legend()
    plt.show 
     
    err = abs(yeval-fex)
    err1 = abs(yeval1-fex)
    err2 = abs(yeval2-fex)
    plt.figure()
    plt.plot(xeval,err1,'co-', label = 'N = 7')
    plt.plot(xeval,err,'go-', label = 'N = 10')
    plt.plot(xeval,err2,'bo-', label = 'N = 18')
    plt.legend()
    plt.show
    
    

    
    
def  eval_lin_spline(xeval,Neval,a,b,f,Nint):

    '''create the intervals for piecewise approximations'''
    xint = np.linspace(a,b,Nint+1)
   
    '''create vector to store the evaluation of the linear splines'''
    yeval = np.zeros(Neval) 
    
    for jint in range(Nint):
        '''find indices of xeval in interval (xint(jint),xint(jint+1))'''
        '''let ind denote the indices in the intervals'''
        '''let n denote the length of ind'''
    
        ind = np.where(np.logical_and(xeval<=xint[jint+1], xeval>=(xint[jint])))
        n = len(ind)
        
        
        
        '''temporarily store your info for creating a line in the interval of 
         interest'''
        a1= xint[jint]
        fa1 = f(a1)
        b1 = xint[jint+1]
        fb1 = f(b1)
        
        for kk in range(n):
            yeval[ind[kk]] = line_maker([a1,b1],[fa1,fb1],xeval[ind[kk]])
        '''use your line evaluator to evaluate the lines at each of the points 
           in the interval'''
        '''yeval(ind(kk)) = call your line evaluator at xeval(ind(kk)) with 
           the points (a1,fa1) and (b1,fb1)'''
    return yeval 

driver()
```

As N is the number of intervals over we are approximating the value of the funciton, the higher N is, the more accurately our linear spline performs, which is as expected. Compared to the experiments in Lab 7 with the global interpolation with uniform nodes, this technique does a lot better with a smaller error for similar orders of N

# 3.3

```python
def find_mi(xint, yint):
    
    n = len(xint)
    mat = np.full((n,n),1/12)
    mat[np.diag_indices_from(mat)] = 1/3
    mat = np.triu(mat,-1)
    mat = np.tril(mat,1)
    y = np.empty(n)
    
    for j in range(1,n-1):
        y[j] = (yint[j+1] - 2*yint[j] + yint[j-1])/(2*(xint[j+1] - xint[j])**2)
    
    invmat = la.inv(mat)
    m = invmat.dot(y)

    return m
```

```python
def cubic_maker(xvals,fvals,mvals,x):
    
    x0 = xvals[0]
    x1 = xvals[1]
    f0 = fvals[0]
    f1 = fvals[1]
    m0 = mvals[0]
    m1 = mvals[1]
    h = x1-x0
    C = (f0/h)-((h*m0)/6)
    D = (f1/h)-((h*m1)/6)
   
    
    yval = (((x1 - x)**3)*m0)/(6*h) + (((x - x0)**3)*m1)/(6*h) + C*(x1 - x) + D*(x - x0)
    
    return yval
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from numpy.linalg import inv 


def driver():
    
    f = lambda x: 1/(1+(10*x)**2)
    a = -1
    b = 1
    
    ''' create points you want to evaluate at'''
    Neval = 100
    xeval =  np.linspace(a,b,Neval)
    
    ''' number of intervals'''
    Nint = 10
    Nint1 = 7
    Nint2 = 18
    
    '''evaluate the linear spline'''
    yeval = eval_cub_spline(xeval,Neval,a,b,f,Nint)
    yeval1 = eval_cub_spline(xeval,Neval,a,b,f,Nint1)
    yeval2 = eval_cub_spline(xeval,Neval,a,b,f,Nint2)
    ''' evaluate f at the evaluation points'''
    fex = np.zeros(Neval)
    for j in range(Neval):
      fex[j] = f(xeval[j]) 
      
    
    plt.figure()
    plt.plot(xeval,fex,'ro-',label='true')
    plt.plot(xeval,yeval1,'cs-',label='approx, N = 7')
    plt.plot(xeval,yeval,'gs-',label='approx, N = 10')
    plt.plot(xeval,yeval2,'bs-',label='approx, N =18')
    plt.legend()
    plt.show 
     
    err = abs(yeval-fex)
    err1 = abs(yeval1-fex)
    err2 = abs(yeval2-fex)
    plt.figure()
    plt.plot(xeval,err1,'co-', label = 'N = 7')
    plt.plot(xeval,err,'go-', label = 'N = 10')
    plt.plot(xeval,err2,'bo-', label = 'N = 18')
    plt.legend()
    plt.show
    
    

    
    
def  eval_cub_spline(xeval,Neval,a,b,f,Nint):

    '''create the intervals for piecewise approximations'''
    xint = np.linspace(a,b,Nint+1)
    yint = np.zeros(Nint+1)
    
    for j in range(Nint):
        yint[j] = f(xint[j])
   
    '''create vector to store the evaluation of the linear splines'''
    yeval = np.zeros(Neval) 
    
    m = find_mi(xint, yint)
    
    for jint in range(Nint):
        '''find indices of xeval in interval (xint(jint),xint(jint+1))'''
        '''let ind denote the indices in the intervals'''
        '''let n denote the length of ind'''
    
        ind = np.where(np.logical_and(xeval<=xint[jint+1], xeval>=(xint[jint])))
        n = len(ind)
        
        
        
        '''temporarily store your info for creating a line in the interval of 
         interest'''
        a1= xint[jint]
        fa1 = f(a1)
        b1 = xint[jint+1]
        fb1 = f(b1)
        
        m0 = m[jint]
        m1 = m[jint+1]
        
        for kk in range(n):
            yeval[ind[kk]] = cubic_maker([a1,b1],[fa1,fb1],[m0,m1],xeval[ind[kk]])
        '''use your line evaluator to evaluate the lines at each of the points 
           in the interval'''
        '''yeval(ind(kk)) = call your line evaluator at xeval(ind(kk)) with 
           the points (a1,fa1) and (b1,fb1)'''
    return yeval 

driver()
```

For N = 10, the highest error is higher in the cubic spline than the linear spline, but for the larger N values, here 18, the highest error is a good amount smaller for the cubic spline
